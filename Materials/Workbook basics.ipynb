 <h2 style='color:red;font-size:30px'> Clustering models <h2>

```python
# K means clustering ------------------------------------------------------------------------------------------- 
```

```python
df = pd.read_csv('./data/income.csv')
```

```python
# Scaling data
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(df[['Age']])
df['Age'] = scaler.transform(df[['Age']])
scaler.fit(df[['Income($)']])
df['Income($)'] = scaler.transform(df[['Income($)']])
df.head()
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=3)
df['Cluster'] = km.fit_predict(df[['Age','Income($)']])
```

```python
df0 = df[df.Cluster==0]
df1 = df[df.Cluster==1]
df2 = df[df.Cluster==2]
plt.figure(figsize=(5,3))
plt.scatter(df0.Age,df0['Income($)'], color='r', label='0')
plt.scatter(df1.Age,df1['Income($)'], color='g', label='1')
plt.scatter(df2.Age,df2['Income($)'], color='b', label='2')
plt.xlabel('Age')
plt.ylabel('Income')
plt.legend()
```

```python
k_rng = range(1, 10)
sse = []
for k in k_rng:
    km = KMeans(n_clusters=k, n_init='auto')
    km.fit(df[['Age','Income($)']])
    sse.append(km.inertia_)
plt.plot(k_rng, sse)
```
