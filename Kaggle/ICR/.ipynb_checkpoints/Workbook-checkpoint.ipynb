```python
# Loading data
import pandas as pd
df_train = pd.read_csv('./data/train.csv')
df_test = pd.read_csv('./data/test.csv')
df_train.head()
```

```python
df_train.drop(['Id'],axis=1,inplace=True)
df_train.head()
```

```python
print(df_train.shape)
print(df_test.shape)
```

```python
df_train.isna().sum()
```

```python
df_train_phase1 = df_train.dropna()
```

```python
df_test_phase1 = df_test.dropna()
```

```python
print(df_train_phase1['EJ'].unique())
print(df_test_phase1['EJ'].unique())
```

```python
train_dummies = pd.get_dummies(df_train_phase1.EJ, dtype='i')
test_dummies = pd.get_dummies(df_test_phase1.EJ, dtype='i')
train_dummies['EJ_A'] = train_dummies['A']
test_dummies['EJ_A'] = test_dummies['A']
train_dummies.head()
```

```python
test_dummies.head(2)
```

```python
df_train_phase2 = pd.concat([df_train_phase1,train_dummies['EJ_A']], axis='columns')
df_test_phase2 = pd.concat([df_test_phase1,test_dummies['EJ_A']], axis='columns')
df_train_phase2.head()
```

```python
df_train_phase2.drop(['EJ'], axis=1, inplace=True)
df_test_phase2.drop(['EJ'], axis=1, inplace=True)
df_train_phase2.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
for column in df_train_phase2.columns:
    if column != 'Class' and column != 'EJ_A':
        scaler.fit(df_train_phase2[[column]])
        df_train_phase2[column] = scaler.transform(df_train_phase2[[column]])
        df_test_phase2[column] = scaler.transform(df_test_phase2[[column]])
df_train_phase2.head()
```

```python
x = df_train_phase2.drop(['Class'], axis=1)
x.head()
```

```python
y = df_train_phase2['Class']
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=5)
```

```python
import tensorflow as tf
from tensorflow import keras

model = keras.Sequential([
    keras.layers.Dense(25, input_shape=(56,)),
    keras.layers.LeakyReLU(alpha=0.2),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(10),
    keras.layers.LeakyReLU(alpha=0.3),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(5),
    keras.layers.LeakyReLU(alpha=0.3),
    keras.layers.Dense(1,activation='sigmoid')
])
model.compile(
    optimizer='adam',
    loss = 'binary_crossentropy',
    metrics = ['accuracy']
) 
model.fit(x_train,y_train, epochs=100)
```

```python
model.evaluate(x_test,y_test)
```

```python
model.fit(x,y, epochs=100)
```

```python
df_test_phase3 = df_test_phase2.drop(['Id'], axis=1)
```

```python
pred = model.predict(df_test_phase3)
last = pd.DataFrame(pred, columns=['Prediction'])
last.head()
```

```python
import numpy as np
last['class_0'] = np.where(last['Prediction']>=0.5, 0, 1)
last['class_1'] = np.where(last['Prediction']>=0.5, 1, 0)
last['Id'] = df_test_phase2['Id']
last.head()
```

```python
last.drop(['Prediction','pred'], axis=1, inplace=True)
```
